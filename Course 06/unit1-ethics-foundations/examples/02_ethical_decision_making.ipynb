# Step 3: Create decision-making flowchart visualization

def create_decision_flowchart():
    """
    Create a flowchart of the ethical decision-making process.
    
    HOW IT WORKS:
    1. Define step positions and properties (colors, sizes, text)
    2. Draw boxes for each step using FancyBboxPatch
    3. Draw arrows connecting steps
    4. Save as high-resolution image
    
    ‚è∞ WHEN to use: After defining framework - see the process visually
    üí° WHY use: Visual representation makes the process easier to understand and remember
    """
    fig, ax = plt.subplots(figsize=(14, 12))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 14)
    ax.axis('off')
    # Define colors for different step types
    start_color = '#2ecc71'  # Green
    process_color = '#3498db'  # Blue
    decision_color = '#f39c12'  # Orange
    end_color = '#9b59b6'  # Purple
    # Define step positions: Create list of step properties (position, size, color, text)
    # Why define positions? Allows us to place steps precisely and draw connections
    steps = [
        {'step': 0, 'name': 'Start', 'x': 5, 'y': 13, 'width': 2, 'height': 0.6, 
         'color': start_color, 'text': 'Start'},  # Start box: Beginning of process
        {'step': 1, 'name': 'Identify Problem', 'x': 5, 'y': 11.5, 'width': 2.5, 
         'height': 0.6, 'color': process_color, 'text': '1. Identify\nProblem'},  # Step 1: Problem identification
        {'step': 2, 'name': 'Gather Info', 'x': 5, 'y': 10, 'width': 2.5, 
         'height': 0.6, 'color': process_color, 'text': '2. Gather\nInformation'},  # Step 2: Information gathering
        {'step': 3, 'name': 'Stakeholders', 'x': 5, 'y': 8.5, 'width': 2.5, 
         'height': 0.6, 'color': process_color, 'text': '3. Identify\nStakeholders'},  # Step 3: Stakeholder identification
        {'step': 4, 'name': 'Apply Frameworks', 'x': 5, 'y': 7, 'width': 2.5, 
         'height': 0.6, 'color': decision_color, 'text': '4. Apply Ethical\nFrameworks'},  # Step 4: Framework application (decision step)
        {'step': 5, 'name': 'Evaluate', 'x': 5, 'y': 5.5, 'width': 2.5, 
         'height': 0.6, 'color': decision_color, 'text': '5. Evaluate\nOptions'},  # Step 5: Option evaluation (decision step)
        {'step': 6, 'name': 'Make Decision', 'x': 5, 'y': 4, 'width': 2.5, 
         'height': 0.6, 'color': process_color, 'text': '6. Make\nDecision'},  # Step 6: Final decision
        {'step': 7, 'name': 'End', 'x': 5, 'y': 2.5, 'width': 2, 'height': 0.6, 
         'color': end_color, 'text': 'End'},  # End box: Completion of process
    ]
    # Draw step boxes: Create visual boxes for each step
    for step in steps:  # Loop through steps: Process each step
        # Create rounded box: Draw styled box for this step
        box = FancyBboxPatch(
            (step['x'] - step['width']/2, step['y'] - step['height']/2),  # Position: Center box at (x, y)
            step['width'], step['height'],  # Size: Use defined width and height
            boxstyle="round,pad=0.1",  # Style: Rounded corners with padding
            edgecolor='black',  # Border: Black outline
            facecolor=step['color'],  # Fill: Use step's color
            linewidth=2,  # Border width: 2 pixels for visibility
            alpha=0.8  # Transparency: 80% opacity for nice look
        )
        ax.add_patch(box)  # Add to plot: Place box on the chart
        
        # Add text label: Place step name/number in box center
        ax.text(step['x'], step['y'], step['text'],  # Position: Center of box
               ha='center', va='center', fontsize=9, fontweight='bold',  # Style: Centered, bold, readable size
               color='white' if step['color'] != '#f39c12' else 'black')  # Text color: White on dark, black on orange
    
    # Draw connecting arrows: Show flow from one step to next
    for i in range(len(steps) - 1):  # Loop through step pairs: Connect consecutive steps
        # Create arrow: Draw arrow from current step to next step
        arrow = FancyArrowPatch(
            (steps[i]['x'], steps[i]['y'] - steps[i]['height']/2),  # Start: Bottom of current step
            (steps[i+1]['x'], steps[i+1]['y'] + steps[i+1]['height']/2),  # End: Top of next step
            arrowstyle='->', mutation_scale=20, linewidth=2, color='black'  # Style: Black arrow, visible size
        )
        ax.add_patch(arrow)  # Add to plot: Place arrow on chart
    
    # Add title: Describe what the flowchart shows
    ax.set_title('Ethical Decision-Making Framework for AI', 
                fontsize=14, fontweight='bold', pad=20)  # Title: Main heading
    
    # Save visualization: Export as high-resolution image
    plt.tight_layout()  # Adjust layout: Prevent label cutoff
    script_dir = os.path.dirname(os.path.abspath(__file__))  # Get directory: Find notebook location
    output_path = os.path.join(script_dir, 'ethical_decision_flowchart.png')  # File path: Save in notebook directory
    plt.savefig(output_path, dpi=300, bbox_inches='tight')  # Save image: High resolution (300 dpi), tight bounds
    print(" Saved: ethical_decision_flowchart.png")  # Success message: Confirm save
    plt.close()  # Close figure: Free memory

# Run the visualization

def create_stakeholder_analysis():
    """Create a stakeholder impact matrix"""
    stakeholders = ['Users', 'Developers', 'Company', 'Society', 'Regulators']
    impact_levels = [9, 7, 8, 6, 7]  # High impact scores
    influence_levels = [5, 8, 9, 4, 9]  # Influence scores
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    # Impact chart
    colors = plt.cm.viridis(np.linspace(0, 1, len(stakeholders)))
    bars1 = ax1.barh(stakeholders, impact_levels, color=colors, alpha=0.8)
    ax1.set_xlabel('Impact Level (1-10)', fontsize=11)
    ax1.set_title('Stakeholder Impact', fontsize=12, fontweight='bold')
    ax1.set_xlim(0, 10)
    ax1.grid(axis='x', alpha=0.3)
    for i, (bar, value) in enumerate(zip(bars1, impact_levels)):
        ax1.text(value + 0.2, i, f'{value}', va='center', fontweight='bold')
    # Influence chart
    bars2 = ax2.barh(stakeholders, influence_levels, color=colors, alpha=0.8)
    ax2.set_xlabel('Influence Level (1-10)', fontsize=11)
    ax2.set_title('Stakeholder Influence', fontsize=12, fontweight='bold')
    ax2.set_xlim(0, 10)
    ax2.grid(axis='x', alpha=0.3)
    for i, (bar, value) in enumerate(zip(bars2, influence_levels)):
        ax2.text(value + 0.2, i, f'{value}', va='center', fontweight='bold')
    plt.tight_layout()
    # Get the directory where this script is located
    script_dir = os.path.dirname(os.path.abspath(__file__))
    output_path = os.path.join(script_dir, 'stakeholder_analysis.png')
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    print(" Saved: stakeholder_analysis.png")
    plt.close()
