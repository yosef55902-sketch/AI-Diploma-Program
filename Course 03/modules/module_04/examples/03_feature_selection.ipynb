## Step 2: Implementation | ÿßŸÑÿÆÿ∑Ÿàÿ© 2: ÿßŸÑÿ™ŸÜŸÅŸäÿ∞

**BEFORE**: We understand the concept.

**AFTER**: We'll see HOW it's implemented!


```python
# Example 1: Feature Selection
# WHY: Keep only important features
# HOW: Select features based on importance

# Simulate feature importance
np.random.seed(42)
n_features = 10
feature_importance = np.random.rand(n_features)
feature_importance = np.sort(feature_importance)[::-1]  # Sort descending

# Select top k features
k = 5
top_features = np.argsort(feature_importance)[-k:][::-1]

print("Example 1: Feature Selection")
print("=" * 60)
print(f"Total features: {n_features}")
print(f"Feature importance scores:")
for i, imp in enumerate(feature_importance):
    marker = "‚úì" if i in top_features else " "
    print(f"  Feature {i}: {imp:.4f} {marker}")
print(f"\nSelected top {k} features: {top_features}")
print(f"\nüí° WHY: Keep only important features")
print(f"üí° HOW: Select features with highest importance")
print(f"üí° AFTER: Simpler, more efficient models!")
```

---

## üìä Visualization | ÿßŸÑÿ™ÿµŸàÿ±


```python
# Visualize feature importance
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['green' if i in top_features else 'gray' for i in range(n_features)]
ax.barh(range(n_features), feature_importance, color=colors, alpha=0.7)
ax.set_yticks(range(n_features))
ax.set_yticklabels([f'Feature {i}' for i in range(n_features)])
ax.set_xlabel('Importance Score', fontsize=12)
ax.set_ylabel('Features', fontsize=12)
ax.set_title('Feature Importance (Green = Selected)', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

print("\nüìä Visualization:")
print("  - Green bars: Selected features")
print("  - Gray bars: Unselected features")
```
