## Step 2: Implementation | ÿßŸÑÿÆÿ∑Ÿàÿ© 2: ÿßŸÑÿ™ŸÜŸÅŸäÿ∞

**BEFORE**: We understand the concept.

**AFTER**: We'll see HOW it's implemented!


```python
# Example 1: Probability Distributions
# WHY: Need to quantify uncertainty
# HOW: Use probability distributions

# Normal distribution (most common in ML)
np.random.seed(42)
mean = 0
std = 1
samples = np.random.normal(mean, std, 1000)

print("Example 1: Probability Distributions")
print("=" * 60)
print(f"Normal distribution: mean={mean}, std={std}")
print(f"Sample mean: {np.mean(samples):.4f}")
print(f"Sample std: {np.std(samples):.4f}")
print(f"\nüí° WHY: Quantifies uncertainty in predictions")
print(f"üí° HOW: Normal distribution models many natural phenomena")
print(f"üí° AFTER: Can provide confidence intervals!")
```

---

## üìä Visualization | ÿßŸÑÿ™ÿµŸàÿ±


```python
# Visualize probability distribution
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(samples, bins=30, density=True, alpha=0.7, edgecolor='black', label='Samples')
x_range = np.linspace(-4, 4, 100)
y_range = (1/(std*np.sqrt(2*np.pi))) * np.exp(-0.5*((x_range-mean)/std)**2)
ax.plot(x_range, y_range, 'r-', linewidth=2, label='Theoretical PDF')
ax.set_xlabel('Value', fontsize=12)
ax.set_ylabel('Probability Density', fontsize=12)
ax.set_title('Normal Distribution', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nüìä Visualization:")
print("  - Histogram: Sampled values")
print("  - Red curve: Theoretical probability density function")
```
